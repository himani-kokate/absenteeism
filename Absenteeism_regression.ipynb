```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
X=df.drop('Absenteeism time in hours',axis=1)
y=df['Absenteeism time in hours']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model=Sequential()
model.add(Dense(64,input_dim=X_train.shape[1],activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=10,validation_split=0.1)
```

```python
predictions=model.predict(X_test).flatten()
mse=np.mean(np.square(predictions-y_test.values))
mae=np.mean(np.abs(predictions-y_test.values))
print(f'Mean Squared Error:{mse}')
print(f'Mean Abaolute Error:{mae}')
```
